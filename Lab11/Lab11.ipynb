```python
import pandas as pd
import pymc as pm
import numpy as np
import arviz as az

file_path = "real_estate_data.csv"
data = pd.read_csv(file_path)

# x1 - suprafata
# x2 - nr camere
# x3 - distanta
# y - pret

x1 = data.iloc[:, 0]
x2 = data.iloc[:, 1]
x3 = data.iloc[:, 2]
y = data.iloc[:, 3]

with pm.Model() as model:
    intercept = pm.Normal("intercept", mu=0, sigma=10)
    beta_x1 = pm.Normal("beta_x1", mu=0, sigma=10)
    beta_x2 = pm.Normal("beta_x2", mu=0, sigma=10)
    beta_x3 = pm.Normal("beta_x3", mu=0, sigma=10)
    sigma = pm.HalfNormal("sigma", sigma=10)

    mu = (intercept +
          beta_x1 * x1 +
          beta_x2 * x2 +
          beta_x3 * x3)

    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y)

    trace = pm.sample(2000, tune=1000, target_accept=0.9)

az.summary(trace, hdi_prob=0.95)
```

Din rezultatele obtinute, putem deduce ca cel mai mare impact il are numarul de camere cu beta_2 = 8.647, dupa distanta fata de centru cu beta_3 = -3.407, dupa suprafata cu beta_1 = 2.445.<br>
Putem interpreta ca pentru fiecare unitate cu care crestea valoarea atributului (1 camera in plus, 1 km mai departe de centru sau 1 mp mai mult), valoarea pretului se va modifica cu beta unitati (o crestere in pret de 8647, -3407, respectiv 2445 euro).<br>
<br>
Coeficientul negativ al distantei se explica prin faptul ca un apartament este mai valoros atunci cand distanta fata de centrul orasului este mai mica. Asadar, cu cat distanta este mai mare, cu atat pretul trebuie sa scada.
